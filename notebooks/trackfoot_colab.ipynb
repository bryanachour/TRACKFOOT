```python
from IPython.display import Video, Image, display
from pathlib import Path
out = sorted(Path('output').glob('run_*'))[-1]
for v in ['annotated.mp4', 'tactical.mp4', 'stacked.mp4']:
    if (out / v).exists():
        display(Video(str(out / v), embed=True, width=900))
for hm in sorted((out / 'heatmaps').glob('*.png'))[:6]:
    display(Image(str(hm)))
```
